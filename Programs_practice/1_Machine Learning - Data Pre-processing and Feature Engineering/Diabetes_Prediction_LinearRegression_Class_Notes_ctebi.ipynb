## Linear Regression on Diabetes Dataset

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_diabetes
```

```python
# Loading diabetes datasets
data = load_diabetes()
```

```python
data
```

```python
print(data['DESCR'])
```

```python
data.data
```

```python
# convert the features into the dataframe
data.keys()
features = pd.DataFrame(data.data)
features
```

```python
features = pd.DataFrame(data.data, columns = data.feature_names)
features.head()
```

```python
target = pd.Series(data.target, name = 'target')
target.head()
```

```python
# based on these features, we will predict the target
```

```python
from sklearn.linear_model import LinearRegression # LinearRegression is a class
```

```python
lr = LinearRegression() #lr is the object of LinearRegression class
```

```python
# Let us start with a simple linear regression model.
# bmi is the feature and target is the label.

X = features['bmi']
```

```python
y = target
```

```python
lr.fit(X,y)
```

```python
X.ndim
```

```python
X.shape
```

```python
X = features['bmi'].values.reshape(442,1)
X.shape
```

```python
lr.fit(X,y)
```

```python
lr.coef_ # This is m (slope)
# what does it tell - Change in y for 1 unit change in x
```

```python
lr.intercept_ # ca value of the best fit line (value of y when x is 0)
```

```python
pred = lr.predict(X)
pred
```

```python
import matplotlib.pyplot as plt
```

```python
## plot between x and y / plot between bmi and target
plt.scatter(X,y, label = 'Actual')
plt.xlabel('BMI')
plt.ylabel('Target')
plt.plot(X,pred,'-r',label = 'Predicted')
plt.legend()
plt.show()
```

```python
## MODEL EVALUATION:

from sklearn.metrics import mean_squared_error
# MSE = average of squared differnce between the actual and the predicted values
```

```python
mse = mean_squared_error(y, pred)
mse # lower the MSE the better is your model
```

```python
r2 = lr.score(X,y) # Returns the R2 values -> Range (0,1) -> higher the better
r2
```
