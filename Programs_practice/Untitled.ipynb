```python
class Student:
    def __init__ (self, name, age, country="India"):
        self.name = name
        self.age = age
        self.country = country
```

```python
s1 = Student("gaurav", 30)
print(s1.name)
s2 = Student("gaurav1", 35, "USA")
print(s2.country)
```
