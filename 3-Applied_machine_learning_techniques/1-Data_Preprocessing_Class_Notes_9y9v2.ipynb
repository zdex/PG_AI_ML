```python
# handling missing data

data = [['John', 25, 65], ['David', 15, 45], ['Alice', 70, 75], ['Bob', 20]]
df = pd.DataFrame(data, columns = ['Name', 'Age', 'Weight'])
```

```python
# dropping the entire row

df1 = df.dropna()
df1
```

```python
# Substituting the missing value with another value, typically, with mean, median or mode.

# Imputation - technique of replacing missing values
```

```python
# Mean = Average (sum of values / number of values)

# Median = Middle Value
# ex: 1, 3, 4, 5, 6, 8, 90
# Mean = 117 / 7 = 17
# Median = 5

# Mode = most frequent value (ususally used with categorical data)
```

```python
df.describe()
```

```python
from sklearn.impute import SimpleImputer
```

```python
help(SimpleImputer)
```

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = 'median')
```

```python
imputer = imputer.fit(df[['Weight']])
```

```python
df['Weight'] = imputer.transform(df[['Weight']])
```

```python
df
```

## handling Categorical Variables

```python
# Ex. Color -> red, green, yellow, blue (Nominal Catgorical variable)
# Ex. Size -> xs, s, m, l, xl (Ordinal Categorical variable)
```

```python
df = pd.DataFrame(data = [['green', 'M', 10.5], ['yellow', 'S', 14.5], ['red', 'S', 15]], columns = ['Color', 'Size', 'Price']) 
```

```python
df
```

```python
df['Size'] = df['Size'].replace({'S':1,'M':2}) # can also use map
```

```python
df
```

## SCALING

```python
# Scaling meaning bringing all numeric features to a similar scale:

#Why?
# one feature doesn't dominate 

# In predicting, house price, you have 2 features: bedroom, sqft size
# values in bedroom: 1,2,3,4
# values in sqft size: 1000, 2000, 5000
```

## 2 common scaling methods:

1. Standardization
2. Normalization

Standardization (Z-score approach): 

2, 6, 10 
Mean = 6 [(2+6+10)/3]
Std Dev = 4 

z = (x - mean)/std

org value -> z-value
2 -> -1
6 -> 0
10 -> 1

Another feature:

10000 -> -1
12000 -> 0
14000 -> 1

Mean = 12000
Std = 2000

Mean = Average
Standard Deviation = SQRT ((x1 - xm)^2 + (x2 - xm)^2.....)/n)

STD -> measurement of spread of the data

y = mx1 + c
x1 -> 10000, 12000, 11000

y = mx2 + c 
x -> 1,2,4,5

y = m1x1 + m2x2 +...

y = 1*10000 + 1+2 + 0
y = 10002

2. Normalization

This technique is used when we need fixed range of scaled values. Range is between 0 to 1.

x_normalized = (x - min)/ (max - min)

x -> x_normalized
10000 -> 0
12000 -> 0.5
14000 -> 1
