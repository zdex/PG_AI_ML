### 📌 Function Syntax in Python

A **function** is a reusable block of code that performs a **specific task**.

---

### 🔹 What is a Function?

- A function is a block of code that performs a specific task and **runs only when it is called**
- You can pass data to a function using **parameters**
- A function can **return data** as a result

---

### 🔹 Syntax

```python
def function_name(param1, param2, ...):
    line1
    line2
    return  # return is the last statement in the function




````python
#Create a Function:
def my_function():
    print("Hello from a function")
    print("good morning")
print("outside the function")
````

<div style="background-color: lightblue; padding: 5px;"><h2>ARBITARY ARGUMENT:- some function can have different no of arguments as it cant be predecided </h2></div>

NOTE: The argument passed are taken as tuple

````python
# Building a function that can find cumulative sum of any number of values
def cum_sum(*num):
    total = 0
    for i in num:
        total += i # total = total + i
    return total
````

````python
cum_sum(2,4,6,8,10,15,20,50.5)
````

# Lambda Functions
 * A lambda function is a small anonymous function.
 * It can have any number of arguments but only one expression.
 * The expression is evaluated and returned.

<b>Basic Syntax</b>
lambda arguments: expression

Why Lambda Function:
   * Good for performing short operations/data manipulations.
   * Using the lambda function can sometime reduce the readability of code.


````python
# Square of a number
square = lambda x: x * x
print(square(3))
````

````python
# Adding Two Numbers
add = lambda x,y: x + y
print(add(2,3))
````

### 📌 map() in Python

`map()` is used to **apply a function to each element** of an iterable.

---

### 🔹 Syntax

```python
map(function, iterable)




````python
#The map() function in Python is a built-in function that applies a given function to each item of an iterable 
#Z(such as a list, tuple, or set) 
#and returns a map object (an iterator) containing the results.
numbers = [1, 2, 3, 4]
squared = map(lambda x: x ** 2, numbers) 
print(list(squared)) 
````

````python
# How to get the same result without the lamba function

numbers = [1, 2, 3, 4]
def squared1(x):
    return x**2
squaredlist = map(squared1, numbers) 
print(list(squaredlist)) 
````

### 📌 `filter()` in Python



`filter()` is used to **select elements** from an iterable based on a **condition**.

---

### 🔹 Syntax

```python
filter(function, iterable)


````python
nums = [1, 2, 3, 4, 5, 6]

result = filter(lambda x: x % 2 == 0, nums)

print(list(result))
````


<div style="background-color: lightblue; padding: 5px;"><h2>Generator in Functions (Python)</h2></div>
A generator function is a special kind of function that returns values one at a time, instead of returning everything at once.

<b>Why use a generator?</b>

Saves memory (does not store all values)

Useful for large data or infinite sequences

Values are produced only when needed

````python
#Normal function: returns all the values at once
def numbers():
    return [1, 2, 3]

print(numbers())
````

````python
#Generator function
#It returns a generator object, not values directly.
def numbers():
    yield 1
    yield 2
    yield 3

print(numbers())
````

````python
def numbers():
    yield 1
    yield 2
    yield 3

for n in numbers():
    print(n)
````
