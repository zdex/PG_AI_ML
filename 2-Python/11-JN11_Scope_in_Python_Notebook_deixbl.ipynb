# Scope in Python

In this notebook, we will learn about:
- Local Scope
- Global Scope



## 1️⃣ Local Scope
Variables created inside a function belong to the local scope.

```python
def my_function():
    x = 10  # Local variable
    print('Inside function:', x)

#my_function()
print(x)  # Uncommenting this will give an error
```

## 2️⃣ Global Scope
Variables created outside all functions belong to the global scope.

```python
y = 20  # Global variable

def show():
   print('Inside function:', y)

show()
print('Outside function:', y)
```

## 3️⃣ Using global Keyword
Use `global` if you want to modify a global variable inside a function.

```python
y = 20  # Global variable

def show():
   #global y
   y=40
   print('Inside function:', y)

show()
print('Outside function:', y)
```
