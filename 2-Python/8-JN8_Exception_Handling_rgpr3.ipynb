<h3> EXCEPTION HANDLING </h3>

```python
#generic exception - try block gets executed without error if there are no exceptions
try:
    print('hello')
    c = 1/10
    print('done')
except:
    print('error in exception Block!!')
print('code after Try Catch')

```

```python
#generic exception - except block gets executed with user defined output for any type of error
try:
    print('hello')
    c = 1/0
    print('done')
except:
    print('error in exception Block!!')
print('code after Try Catch')

```

```python
# apply multiple types of exception handling
try:
    num = int(input('enter a number:'))
    c =100/num
    print("100 divide by",num,":",c)
    print('done')
except ZeroDivisionError as e:
    print(e)
    print('Zero is an invalid number')
except ValueError:
    print('Value error!!')
    print('text is a bad input')
print('Awesome!!')
```

### 📌 Custom Error (Custom Exception) in Python

A **custom error** is a user-defined exception that is raised when a specific condition is met.

---

### 🔹 Scenario

A company has a policy:
- If the **employee age is less than 18**, raise an exception



```python
try:
    age = int(input("Enter the age"))
    if age<18:
        raise ValueError
except:
        print('you are disqualified')
    
        
```

```python
# we can also create a Custom Class:
class ageexception(Exception):
   pass
```
