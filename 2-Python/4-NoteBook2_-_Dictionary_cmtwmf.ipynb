<h1>Dictionaries<h1>

```python
#Dictionaries are used to store data values in key:value pairs.
# A dictionary is a collection which is ordered*, changeable and do not allow duplicates in keys.
```

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
print(thisdict)
len(thisdict)
```

```python
# Keys cant be duplicated – duplicates are ignored
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964,
  "year": 2024
}
print(thisdict)
```

```python
# values can be duplicated 
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964,
  "year1": 1964
}
print(thisdict)
```

```python
# The values of item can be of any data types
thisdict = {
  "brand": "Ford",
  "model": [1,2,3],
  "year": {19,64},
  "year1": (1,2,3),
    "dict":{
  "brand": "Ford",
  "model": [1,2,3],
  "year": {19,64},
  "year1": (1,2,3)
}
}
print(thisdict)
```

```python
#dictionary keys can have keys of various data types as long as those data types are immutable.
thisdict = {
  1: "Ford",
  [1,2,3]: [1,2,3],
  {19,64}: {19,64},
  (1,2,3): (1,2,3)
}
thisdict1 = {
  1: thisdict,
  #[1,2,3]: [1,2,3],
  #{19,64}: {19,64},
  (1,2,3): (1,2,3)
}
print(thisdict)
```

```python
#Accessing Items/values- Using Index or Get method
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
x = thisdict["model"]
y = thisdict.get("year")
print(x,y)
```

```python
#Get Values
#The values() method will return a list of all the values in the dictionary.
x = thisdict.values()
print(x)
```

```python
#Change the value of the Items
thisdict["model"] ='audi'
print(thisdict)
```

```python
# update() method will update the dictionary. 
# update() method will update if the key is already present otherwise will update the dict
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.update({"year": 2020})
thisdict.update({"color": "red"})
print(thisdict)
```

```python
#Adding Items
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict["color"] = "red"
print(thisdict)

```
