## Plot using Matplotlib vs Seaborn

Seaborn does the calculation automatically

The tips dataset has many bills for each day
you say:
👉 x = day, y = total_bill
- Seaborn automatically calculates averages for you

- Matplotlib won’t do that—you must calculate manually.

Black line = how much the data varies around that average
Example:

On Sunday, some people pay ₹200

Some pay ₹2000

The average might be ₹800,
but because values are very spread out → big black line.


```python
# PLotting avg bill amount by Day of the Week

# using Matplotlib

grouped = tips.groupby("day")["total_bill"].mean()
grouped
plt.bar(grouped.index, grouped.values)
plt.title("Average Bill Amount by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```python
# using seaborn

sns.barplot(x="day", y="total_bill", data=tips)
plt.title("Average Bill by Day")
plt.show()
```

```python
# using seaborn but changing the aggregation for calculating total bill for each day
import numpy as np

sns.barplot(x="day", y="total_bill", data=tips, estimator = np.sum )
plt.title("Average Bill by Day")
plt.show()
```

### Seaborn vs. Matplotlib

#### Matplotlib
- Low-level plotting library
- Highly **flexible and customizable**
- Can create almost **any type of plot**
- Requires **more code**
- Has a **steeper learning curve**
- Limited default styles and color palettes
- Users often need to manually customize plots

---

#### Seaborn
- High-level interface built on **Matplotlib**
- Designed for **statistical graphics**
- Requires **less code** to create plots
- Provides **built-in statistical functions**
- Makes complex statistical visualization easier
- Offers **better default styles and color palettes**
- Produces attractive plots with **minimal customization**

---

### Key Difference (Summary)
 **Seaborn** automatically handles aggregation and styling,
while **Matplotlib** requires manual data preparation but offers more control


```python
import matplotlib.pyplot as plt
import pandas as pd

# Create a DataFrame
data = {
    'Fruit': ['Apple', 'Banana', 'Cherry', 'Orange', 'Peach', 'Plum', 'Grape'],
    'Weight': [150, 120, 10, 130, 140, 50, 5],
    'Sweetness': [7, 8, 9, 6, 8, 7, 6],
    'Category': ['A', 'B', 'A', 'B', 'C', 'C', 'A']
}

df = pd.DataFrame(data)

# Map categories to colors
colors = {'A': 'red', 'B': 'blue', 'C': 'green'}

# Plot using Matplotlib
plt.figure(figsize=(5, 5))
for category, color in colors.items():
    subset = df[df['Category'] == category]  #Filter Data for Each Category e.g A for the first loop, B for second etc
    plt.scatter(subset['Weight'], subset['Sweetness'], c=color, label=category, s=100)  # s is the size of the markers

# Add title and labels
plt.title('Weight vs Sweetness (Matplotlib)')
plt.xlabel('Weight')
plt.ylabel('Sweetness')
plt.legend()
plt.savefig("abc.png")
# Show plot
plt.show()
```

```python
import seaborn as sns
import pandas as pd

# Create a DataFrame
data = {
    'Fruit': ['Apple', 'Banana', 'Cherry', 'Orange', 'Peach', 'Plum', 'Grape'],
    'Weight': [150, 120, 10, 130, 140, 50, 5],
    'Sweetness': [7, 8, 9, 6, 8, 7, 6],
    'Category': ['A', 'B', 'A', 'B', 'C', 'C', 'A']
}

df = pd.DataFrame(data)

# Plot using Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Weight', y='Sweetness', hue='Category', palette='Set1', s=100) # s is the size of the markers

# Add title
plt.title('Weight vs Sweetness (Seaborn)')

# Show plot
plt.show()
```
