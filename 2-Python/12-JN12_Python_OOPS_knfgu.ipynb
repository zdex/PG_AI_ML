# 🧠 Python OOPS – Classes, Objects, Methods, Constructors & Static Methods



## 📌 What is OOPS?
**OOPS (Object Oriented Programming System)** means writing programs using **objects**.

Each object has:
- **Data** (name, age, balance)
- **Actions** (study, drive, withdraw)

## 🧱 Class – Blueprint
A **class** is a blueprint or template.

🏠 Blueprint of a house → Class
🏠 Actual house → Object

Class can have its properties (attributes) -> Size, Bedrooms, Balcony, Terrace

Object can have value of the attributes -> Size = 2000sft, Bedrooms = 4, Balcony = 2, Terrance = 1

Objects cab have methods to perform some actions -> clean my house, paint my house etc

```python

class Student:
   pass
   # The class above is empty. It just defines a structure.
```

## 🧍 Object – Real World Entity
An **object** is created from a class.

Each object is independent.

```python

s1 = Student()
s2 = Student()

print(s1)
print(s2)
```

## ⚙️ Constructor – __init__()
A **constructor** is a special method:
- Runs automatically when an object is created
- Used to assign values


### 🎒 Real‑life example
When a student joins a school, we record name and age immediately.


```python
# Student class having 3 object attributes (name, age, country)
# self means the the attributes are of individual objects/instances.

class Student:
   def __init__(self,name,age,country='india'): # classes can have default attributes
       self.name = name
       self.age = age
       self.country = country
```

```python
s1 = Student("Rahul", 15)
```

```python
dir(Student)
```

```python
# calling object attributes
s1.name
```

```python

s1 = Student("Rahul", 15)
s2 = Student("Anita", 14)
s3 = Student("mark", 14,"USA")

print(s1.name, s1.age)

print(s3.name, s3.age,s3.country)
```

## 🤔 What is self?
`self` refers to the **current object**.

It helps Python know **which object's data** is being used.


## 🔧 Methods – Actions of a Class
Methods are functions written inside a class.


```python

class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        print("My name is", self.name)
        print("My age is", self.age)

    def study(self):
        print( self.name," is studying")
      
    def sportsDay(self):
        print( self.name," is participating")
```

```python

s1 = Student("Joe", 15)
s1.introduce()
s1.study()
s1.sportsDay()

```

```python
s2 = Student("Neha", 15)
s2.introduce()
s2.sportsDay()
```

## 🏦 Example – Bank Account

```python

class BankAccount:
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print("Deposited:", amount)

    def check_balance(self):
        print("Current Balance:", self.balance)
```

```python

acc = BankAccount("Amit", 5000)
acc.check_balance()
acc.deposit(2000)
acc.check_balance()
acc.deposit(5000)
acc.check_balance()
```

## 📌 Static Variable
A **static Variable**:
- Belongs to the class, not to any one object
- Is shared by all objects
- Same value for every object

🎓 Real-life example

In a school:
School name is the same for all students
Passing marks are same for all students

So these should be static variables.


```python

class Student:
    school_name = "XYZ Public School"   # static variable

    def __init__(self, name, marks):
        self.name = name        # instance variable
        self.marks = marks
```

```python
s1 = Student("Rahul", 85)
s2 = Student("Anita", 92)

print(s1.school_name)
print(s2.school_name)
print(Student.school_name) # static variables can be accessed via classname

```

## 📌 Static Method

Imagine a school rule:

👉 “If marks ≥ 40, the student is PASS, otherwise FAIL”

🔑 Important:

This rule is the same for all students
It does not depend on any one student object
It is just a rule
So this rule should NOT belong to one student
➡️ It should belong to the Student class itself

That is exactly why we use a static method.

👉 A static method is:

A method that belongs to the class

Does not use self

Represents rules / calculations / utility logic

```python
class Student:
    pass_marks = 40

    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    @staticmethod
    def check_result(marks):
        if marks >= Student.pass_marks:
            return "PASS"
        else:
            return "FAIL"
```

```python
print(Student.check_result(85))
print(Student.check_result(32))
s1 = Student("amit",50)
print(s1.check_result(50))
s1.name
```
