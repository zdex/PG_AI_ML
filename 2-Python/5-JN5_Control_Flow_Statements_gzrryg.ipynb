<div style="background-color: lightblue; padding: 5px;"><h2>If Else</h2></div>

```python
# Note: Indentation is very important
a = 10
b = 20
if a < b:
    print(a)
```

```python
# If else: Note: Indentation is very important
a = 30
b = 33
if b < a:
    print('a is greater than b') 
elif a == b:
    print("a and b are equal")
else:
    print("b is greater than a")
```

```python
# If else: Example 2 with AND/OR
a = 200
b = 33
c = 500
if a > b and c > a:
  print("Both conditions are True")

```

```python
# If else: Example with Nested If
x = 5

if x > 10:
  print("Above ten,")
  if x > 20:
    print("and also above 20!")
  else:
    print("but not above 20.")
```

```python
#Pass Statement: if statements cannot be empty,
                 #but if you for some reason have an if statement with no content, 
                 #put in the pass statement to avoid getting an error.
a = 33
b = 200

if b > a:
   pass
print('hey')
```

<div style="background-color: lightblue; padding: 5px;"><h2>While Loop</h2></div>

```python
i = 1
while i < 6:
  print(i)
  i += 1    
print('out of while')
```

```python
# While Loop With Else Statement
i = 1
while i < 6:
  print(i)
  i += 1
else:
  print("i is no longer less than 6")
```

<div style="background-color: lightblue; padding: 5px;"><h2>For Loop</h2></div>

```python
# For Loops: Example 1 - Loop through list
fruits = ["apple", "banana", "cherry"]
for x in fruits:
    print(x)
```

```python
# For Loops: Example 2 - Loop through String
for x in "banana":
  print (x) 
```

```python
#Break in For Loop
lst =[1,2,3,4,5]
for x in lst:
  if x == 3:
    continue
  print(x)

```

```python
#continue in For Loop:Skipping Specific Characters in a String

#print all the letters other than vowels

string = "hello world"
vowels = "aeiou"

for x in string:
    if x in vowels:
        continue  # Skip vowels
    print(x)
```

<div style="background-color: lightblue; padding: 5px;"><h2>The range() Function</h2></div>

```python
# To loop through a set of code a specified number of times, we can use the range() function,
# The range() function returns a sequence of numbers, starting from 0 by default, and increments by 1 (by default), and ends at a specified number.
```

```python
#else in For loop

for x in range(6):
  print(x)
```

```python
# Range Parameters

for x in range(3):  # 0 by default
  print(x)
print('Next')
for x in range(2, 6): # [2,3,4,5]
  print(x)
print('Next')
for x in range(2, 30, 5):
  print(x)
```

<div style="background-color: lightblue; padding: 5px;"><h2>Structural Pattern Matching</h2></div>


```python
value = 42

match value:
    case 0:
        print("Zero")
    case 3.14:
        print("Pi")
    case 42:
        print("The answer to life, the universe, and everything")
    case "Some Name":
        print("Hello")
    case _:
        print("Something Else")
```
