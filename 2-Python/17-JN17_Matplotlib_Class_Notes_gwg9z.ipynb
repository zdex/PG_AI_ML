## 2️⃣ Importing Matplotlib

```python
import matplotlib.pyplot as plt #matplotlib.pyplot is a module in the Matplotlib library used to create graphs easily.
```

## 3️⃣ Line Plot

```python

x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()

```

## 4️⃣ Bar Chart
A Bar Chart compares categorical data.

X-axis shows categories

Height represents value or count

```python
# Adjusting the size of the plot

names = ["Amit", "Bob", "Charan", "Devang"]
marks = [70, 85, 60, 90]
plt.figure(figsize=(4, 2))   # width=4, height=2
plt.bar(names, marks, color = 'green') # put the parameters in the sequence 1) x axis 2) y axis
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
# Giving fixed Y axis range and ticks

departments = ['IT', 'HR', 'Sales']
employee_count = [25, 6, 8]

plt.bar(departments, employee_count)
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.title("Employees per Department")
plt.ylim(0,100) 
plt.yticks([0,25,50,75,100])
plt.show()
```

## 5️⃣ Scatter Plot

```python

x = [1, 2, 3, 4, 10]
y = [5, 7, 8, 6, 9]

plt.scatter(x, y)
plt.plot(x, y)
plt.title("Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
Product = ['P1', 'P2', 'P3']
Sales = [80, 200, 120]

plt.pie(Sales, labels = Product)
plt.title("Sales by Product")
plt.show()
```

```python
# Showing multiple series in a single plot

import matplotlib.pyplot as plt
import numpy as np

# Example data
X = np.array([1, 2, 3, 4, 5, 6])
y_actual = np.array([3, 6, 8, 7, 12, 14])
y_pred = np.array([2.8, 5.1, 6.9, 9.2, 10.8, 13.1])

plt.figure(figsize=(8,5))

# Actual values (scatter)
plt.scatter(X, y_actual, label="Actual", color = 'blue', marker='o')

# Predicted values (line)
plt.plot(X, y_pred, label="Predicted", color = 'red')

plt.title("Actual vs Predicted")
plt.xlabel("X")
plt.ylabel("Y")

plt.legend()


plt.show()
```
