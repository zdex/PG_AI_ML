<div style="background-color: lightblue; padding: 5px;"><h2> Random Module</h2></div>

<h1> Random Module </h1>

* The random module in Python is used to generate random numbers and perform operations.
* It includes functions to generate random integers, floating-point numbers, random choices from sequences, and more.

```python
import random as r
```

```python
print(dir(random))
```

```python
#Random Integer
random_integer = r.randint(1, 10)  # Random integer between 1 and 10 (inclusive)
print(random_integer)
```

```python
#Random Floating-Point Number
random_float = random.random()  # Random float between 0 and 1
print(random_float)
```

```python
#Random Choice from a Sequence
fruits = ['apple', 'banana', 'cherry', 'date']
random_fruit = random.choice(fruits)  # Randomly selects one fruit from the list
print(random_fruit)
```

```python
#Shuffling a Sequence
numbers = [1, 2, 3, 4, 5]
random.shuffle(numbers)  # Shuffles the list in place
print(numbers)
```

<div style="background-color: lightblue; padding: 5px;"><h2> JSON Module</h2></div>

```python
#The json module in Python helps you work with JSON data.
import json
```

```python
dir(json)
```

```python
#Converting Python Objects to JSON (Serialization)
#Dictionary to JSON
person = {
    "name": "Alice",
    "age": 30,
    "city": "New York"
}
json_string = json.dumps(person)
print(json_string)
print(type(json_string))
```

```python
#Example: JSON to Dictionary
json_string = '{"name": "Alice", "age": 30, "city": "New York"}'
person = json.loads(json_string)
print(person)
print(type(person))
```
