<b><h1>Sets</h1>
* Sets are used to store multiple items in a single variable.
* A set is a collection which is unordered, unchangeable*, and unindexed.
* Elements in the set are immutable but you can add and delete elements
* Can't have Duplicate Values
* Use {} to Create Set

```python
s={1,2,3,4,"hi"}
print(s)
s1={2,3,4,"hi",0,True,2,7,False,0} #sets cant have duplicates, Strings are Immutable
#True is treated as 1
#False is treated as 0
print(s1)
```

```python
# length
len(s)

```

```python
# type
type(s)
```

```python
#ACCESS SET ITEMS
#We cant access the set item as they aren’t indexed
#BUT
#you can loop through:
for x in s:
    print(x)
```

```python
#Elements in the set are immutable
myset={1,2,3,'a',(1,2)}
myset[1]='hi'
```

```python
#Can add item: 

#ADD():
s.add("hey")
print(s)
```

```python
#Can Remove Items:
thisset = {"apple", "banana", "cherry"}
thisset.remove("banana")
print(thisset)
```
