<h1>*****************FILE HANDLING IN PYTHON***************</h1>

Files are units of digital data storage that contain information such as text, images, videos, programs, or any other type of data.

<B>Types of files:</B>
1.	Text Files: Contain plain text and are often used for documents, scripts, and configuration files.
2.	Binary Files: Contain data in a format that is not directly human-readable, such as executable programs, images, videos, and databases.

<B>File Handling means:</B>
	Opening the file
	Performing some operations on it
	Closing the file

<B>Why file Handling?</B>

	File handling is essential in Python, as in any programming language, because it allows you to work with external files stored on your computer or network.
	Whenever you take input from the user, the values inputted by the user is stored in random access memory/ volatile memory, so when the app is closed, the values are lost, so if you want to store the data input the user – there are two ways:
-	File Handling
-	Database

File handling has disadvantages over Database like:

	Data consistency
	Data Redundancy
For smaller app, use file handling
For large app, use DataBase

### 📌 Context Manager in Python

A **context manager** is used to **manage resources automatically**.  
It ensures that a resource is **properly opened and closed**.
Explanation:

open("test1.txt", "r") opens the file in read mode

with ensures the file is automatically closed after use

as f stores the file object in variable f

f.read() reads the file content into data

👉 The file is closed automatically when the with block ends, even if an error occurs.

```python
#Example WITHOUT Context Manager ❌
f = open("test1.txt", "r")
data = f.read()
f.close()
```

```python
#Example WITH Context Manager
with open("test1.txt", "r") as f:
    data = f.read()
    print(data)
```
