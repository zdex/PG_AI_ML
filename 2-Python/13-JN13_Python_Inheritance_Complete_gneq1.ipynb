
# Inheritance in Python 🧬

Inheritance allows one class (child) to acquire properties and methods of another class (parent).
It helps in **code reuse**, **readability**, and **maintainability**.



## 1️⃣ Single Inheritance
A child class inherits from **one** parent class.


```python

class Parent:
    def show_parent(self):
        print("This is Parent class")

class Child(Parent):
    def show_child(self):
        print("This is Child class")

obj = Child()
obj.show_parent()
obj.show_child()
```


## 2️⃣ Multilevel Inheritance
A class is derived from another derived class.


```python

class GrandParent:
    def gp_method(self):
        print("Grandparent method")

class Parent(GrandParent):
    def parent_method(self):
        print("Parent method")

class Child(Parent):
    def child_method(self):
        print("Child method")

obj = Child()
obj.gp_method()
obj.parent_method()
obj.child_method()
```


## 3️⃣ Multiple Inheritance
A child class inherits from **more than one parent class**.


```python

class Father:
    def father_method(self):
        print("Father method")

class Mother:
    def mother_method(self):
        print("Mother method")

class Child(Father, Mother):
    def child_method(self):
        print("Child method")

obj = Child()
obj.father_method()
obj.mother_method()
obj.child_method()
```


## 4️⃣ Hierarchical Inheritance
Multiple child classes inherit from **one parent class**.


```python

class Parent:
    def common_method(self):
        print("Common Parent method")

class Child1(Parent):
    pass

class Child2(Parent):
    pass

obj1 = Child1()
obj2 = Child2()

obj1.common_method()
obj2.common_method()
```


## 5️⃣ Hybrid Inheritance
Combination of two or more types of inheritance.


```python

class A:
    def method_a(self):
        print("Class A")

class B(A):
    def method_b(self):
        print("Class B")

class C(A):
    def method_c(self):
        print("Class C")

class D(B, C):
    def method_d(self):
        print("Class D")

obj = D()
obj.method_a()
obj.method_b()
obj.method_c()
obj.method_d()
```


## 📝 Summary
- **Single** → One parent, one child  
- **Multilevel** → Parent → Child → Grandchild  
- **Multiple** → Multiple parents, one child  
- **Hierarchical** → One parent, many children  
- **Hybrid** → Combination of types
